```python
import pandas as pd
import random
def marksdata(filename,ro,n):
    for i in range (0,n):
        rd=random.randint(0,100)
        df=pd.read_csv(filename)
        ro=ro+1
        df.loc[n]=ro,rd
        df.to_csv(filename,index=False)
    return df
marksdata('FileData/marks.csv',221710310000,2000)
df=marksdata('FileData/marks.csv',221710310000,2000)
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
k=df.values
roll=[]
mark=[]
a=[]
b=[]
c=[]
f=[]
h=0
for i in range (0,len(k)):
    x=df.values[i][0]
    roll.append(x)
for i in range (0,len(k)):
    y=df.values[i][1]
    mark.append(y)
for i in range(0,len(mark)):
    if(mark[i]>90):
        x=df.values[i]
        a.append(x)
    elif mark[i]>80 and mark[i]<90:
        x=df.values[i]
        b.append(x)
        print(mark[i],i)
    elif mark[i]>70 and mark[i]<80:
        x=df.values[i]
        c.append(x)  
        print(mark[i],i)
    elif mark[i]>0 and mark[i]<40:
        x=df.values[i]
        f.append(x)
        print(mark[i],i)
print(len(a),len(b),len(c),len(f),h)
```

```python
labels=['Grade A','Grade B','Grade C','Grade F']
sizes=[len(a),len(b),len(c),len(f)]  
explode=(0.1,0,0,0)

fidl,axl=plt.subplots()
axl.pie(sizes,explode=explode,labels=labels,shadow=True,startangle=90)
axl.axis('equal')
plt.show()
```

```python
w=1/5
marks=[len(a),len(b),len(c),len(f)]
roll=['A','B','C','F']
plt.bar(roll,marks,w)
plt.xlabel('Numbers of students')
plt.ylabel('Grades')
plt.title('Total')
plt.show
```
