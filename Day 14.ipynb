Define a funtion to add new row data to .cvs files

```python
import pandas as pd
def AddrowCSV(filename,rd):
    df=pd.read_csv(filename)
    df.loc[len(df)]=rd
    df.to_csv(filename,index=False)
    return df
rd=[9,'874554','sadsasa',3535,4335,453,9513,8545,85386,521,8463,6863]
AddrowCSV('FileData/income.csv',rd)
```

```python
def updateROWCSV(filename,ri,rd):
    df=pd.read_csv(filename)
    df.loc[ri]=rd
    df.to_csv(filename,index=False)
    return df
data=[10,'87454','bad ck',3535,4335,453,9513,8545,85386,521,8463,6863]
updateROWCSV('FileData/income.csv',9,data)
#readCSV('FileData/income.csv')
```

```python
def dela(filename,rd):
    df=pd.read_csv(filename)
    df=df.drop(rd)
    df.to_csv(filename,index=False)
    return df
dela('FileData/income.csv',6)
```

```python
import random
def marksdata(n,filename):
    with open(filename,'w') as f:
        for i in range (0,n):
            mark=random.randint(0,101)
            f.write(str(mark)+'\n')
        return
marksdata(1300,'FileData/marks.cvs')
```
